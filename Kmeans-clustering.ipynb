```
# 1) Load the image data

from sklearn.metrics import pairwise_distances_argmin
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
img1 = io.imread('panoramic.jpeg')
img2 = io.imread('biomedical.jpg')
```

```
# 2) Define a python function for k-means clustering algorithm
def k_means_clustering(img, n_clusters, rseed=2):
    rng = np.random.RandomState(rseed)
    i = rng.permutation(img.shape[0])[:n_clusters]
    j = rng.permutation(img.shape[1])[:n_clusters]

    centers = list()
    for it in range(n_clusters):
        centers.append(img[i[it],j[it]])
    centers = np.array(centers)
    
    while True:
        labels = list()
        for it in range(img.shape[0]):
            label = pairwise_distances_argmin(img[it], centers)
            labels.append(label)
        labels = np.array(labels)

        new_centers = list()
        for n in range(n_clusters):
            sum_val = np.zeros(3)
            freq = 0
            for it in range(img.shape[0]):
                for jt in range(img.shape[1]):
                    if (labels[it][jt] == n):
                        sum_val = sum_val + img[it][jt]
                        freq = freq + 1
            new_centers.append((sum_val/freq))

        new_centers = np.array(new_centers)
        
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels
```

```
# 3) Apply k-means clustering to each pixel of the image using k=2

centers_img1_k2, labels_img1_k2 = k_means_clustering(img1, 2)

centers_img2_k2, labels_img2_k2 = k_means_clustering(img2, 2)
```

```
# 4) Apply k-means clustering to each pixel of the image using k=3

centers_img1_k3, labels_img1_k3 = k_means_clustering(img1, 3)

centers_img2_k3, labels_img2_k3 = k_means_clustering(img2, 3)
```

```
# 5) Visualize the clustering results for k =2 and k=3

from PIL import Image

def create_visualization(img, centers, labels, k):
    new_c = list()
    for i in range(k):
        nwc = list()
        for j in range(centers[i].size):
            nwc.append(int(centers[i][j]))
        new_c.append(nwc)
    new_c = np.array(new_c)

    new_img = list()
    for i in range(img.shape[0]):
        temp = list()
        for j in range(img.shape[1]):
            if labels[i][j] == 0:
                temp.append(new_c[0])
            elif labels[i][j] == 1:
                temp.append(new_c[1])
            elif k==3:
                if(labels[i][j] == 2):
                    temp.append(new_c[2])
        new_img.append(temp)

    new_img = np.array(new_img)
    new_img = np.uint8(new_img)

    return new_img

new_img1_k2 = create_visualization(img1, centers_img1_k2, labels_img1_k2, 2)
new_img1_k3 = create_visualization(img1, centers_img1_k3, labels_img1_k3, 3)

new_img2_k2 = create_visualization(img2, centers_img2_k2, labels_img2_k2, 2)
new_img2_k3 = create_visualization(img2, centers_img2_k3, labels_img2_k3, 3)

img = Image.fromarray(new_img1_k2, 'RGB')
img.save('panoramic_k2.jpg')
img.show()

img = Image.fromarray(new_img1_k3, 'RGB')
img.save('panoramic_k3.jpg')
img.show()

img = Image.fromarray(new_img2_k2, 'RGB')
img.save('biomedical_k2.jpg')
img.show()

img = Image.fromarray(new_img2_k3, 'RGB')
img.save('biomedical_k3.jpg')
img.show()
```
